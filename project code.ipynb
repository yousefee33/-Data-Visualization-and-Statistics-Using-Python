```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
DF=pd.read_csv("Diabetes Dataset.csv")
```

```python
DF
```

# 1. Correlation Matrix Plot

```python
plt.figure(figsize=(10, 8))
corr = DF.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# 2. Scatter Plot (BloodPressure vs BMI)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=DF, x='BloodPressure', y='BMI')
plt.title('Scatter Plot: BloodPressure vs BMI')
plt.xlabel('Blood Pressure')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

# 3. Distribution Plots for specified features

```python
features = ['Pregnancies', 'SkinThickness', 'Age']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.histplot(DF[feature], kde=True, bins=20, color='blue', edgecolor='black')
    plt.title(f'Distribution Plot for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

# 4. Pie Chart for Outcome (Diabetes or Not)

```python
plt.figure(figsize=(6, 6))
outcome_counts = DF['Outcome'].value_counts()
outcome_labels = ['No Diabetes', 'Diabetes']
plt.pie(outcome_counts, labels=outcome_labels.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title("Pie Chart: Outcome (Diabetes or Not)")
plt.show()
```

# 5. Boxplots for specified features

```python
boxplot_features = ['SkinThickness', 'DiabetesPedigreeFunction', 'BloodPressure']
for feature in boxplot_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=DF, x=feature)
    plt.title(f'Boxplot for {feature}')
    plt.xlabel(feature)
    plt.show()
```

# 6. Statistical Calculations

```python
central_tendency = DF.describe().T[['mean', '50%', 'std']]
central_tendency.rename(columns={'50%': 'median'}, inplace=True)
central_tendency['mode'] = DF.mode().iloc[0]
print("Central Tendency (Mean, Median, Mode):")
print(central_tendency)
```

# Dispersion: Standard Deviation, Variance

```python
dispersion = DF.std()
variance = DF.var()
dispersion_df = pd.DataFrame({'Standard Deviation': dispersion, 'Variance': variance})
print("\nDispersion (Standard Deviation, Variance):")
print(dispersion_df)
```

# Minimum and Maximum Values

```python
min_max = pd.concat([DF.min(), DF.max()], axis=1)
min_max.columns = ['Min', 'Max']
print("\nMinimum and Maximum Values:")
print(min_max)
```
